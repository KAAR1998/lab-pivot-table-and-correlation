#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
df_avg_sal_num_emp = df.pivot_table(index=['Department', 'Title'], aggfunc=({"Salary":"mean", "Name":"count"}))

df_avg_sal_num_emp = df_avg_sal_num_emp.rename(columns={"Name":"Employee Count"})

display(df_avg_sal_num_emp)
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
# your answer here
sal_med_min_max = df.pivot_table(index = ['Department', 'Title'], values = ['Salary'], aggfunc =['median', 'min', 'max', 'std'])

display(sal_med_min_max)
```

#### Based on your comments, fill the missing values with an appropriate value.

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
# your answer here
salary_by_dept = df.pivot_table(index='Department', aggfunc=({"Salary" : "sum", "Name" : "count"}))

salary_by_dept = salary_by_dept.rename(columns={"Name":"Employee Count", "Salary":"Salary Spend"})

salary_by_dept.sort_values(by='Employee Count', ascending=False)
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

```python
# your answer here

we_ms = df.pivot_table(index=['Department', 'Title'], aggfunc=({"Salary" : "mean", "Years" : "sum"}) )

we_ms = we_ms.rename(columns={"Salary":"Mean Salary", "Years":"Total Years of Work Experience"})

display(we_ms)
```
